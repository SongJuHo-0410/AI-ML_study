```python
#라이브러리 불러오기
from sklearn.tree import DecisionTreeClassifier   
from sklearn.datasets import load_iris    
from sklearn.model_selection import train_test_split

#iris 데이터 불러오기, 학습셋-데이터셋 나누기
iris_data=load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, test_size=0.2, random_state=10)

#DecisionTreeClassifier 생성
clf= DecisionTreeClassifier(random_state=1) 

#결정트리 학습
clf.fit(X_train, y_train)

#정확도
print("훈련 세트 정확도: {:.3f}".format(clf.score(X_train, y_train)))
print("테스트 세트 정확도: {:.3f}".format(clf.score(X_test, y_test)))
```

```python
#라이브러리 불러오기
import matplotlib.pyplot as plt
from sklearn import tree

#g맷플롯립 이용하여 시각화
plt.figure(figsize = (20,15))
tree.plot_tree(clf,
                class_names = iris_data.target_names,
                feature_names=iris_data.feature_names,
                impurity=True, filled =True,
                rounded = True, fontsize = 16)

plt.show()
```

```python
#정확도
print("훈련 세트 정확도: {:.3f}".format(clf.score(X_train, y_train)))
print("테스트 세트 정확도: {:.3f}".format(clf.score(X_test, y_test)))
```

```python
#DecisionTreeClassifier 생성
clf= DecisionTreeClassifier(max_depth=4,  random_state=1) 

#결정트리 학습
clf.fit(X_train, y_train)

#정확도
print("훈련 세트 정확도: {:.3f}".format(clf.score(X_train, y_train)))
print("테스트 세트 정확도: {:.3f}".format(clf.score(X_test, y_test)))
```

```python
#DecisionTreeClassifier 생성
clf= DecisionTreeClassifier(max_depth=4,  random_state=1) 

#결정트리 학습
clf.fit(X_train, y_train)

#g맷플롯립 이용하여 시각화
plt.figure(figsize = (20,15))
tree.plot_tree(clf,
                class_names = iris_data.target_names,
                feature_names=iris_data.feature_names,
                impurity=True, filled =True,
                rounded = True, fontsize = 16)

plt.show()
```

```python
#정확도
print("훈련 세트 정확도: {:.3f}".format(clf.score(X_train, y_train)))
print("테스트 세트 정확도: {:.3f}".format(clf.score(X_test, y_test)))
```
